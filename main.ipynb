###  Data Visualization:
       1-Raw data
       2-After filtration using High pass filter
       3-After RMS.
       4-After Envelope

```python
emg_signal  = data['emg'][7] 
window_size = len(emg_signal) / 2000 
w_length=64
zooming_from_to = (170,180)
sample_frequency = 2000
cutoff_frequency = 20
#Before filtration
plt.figure(1)
plt.title('Before filtration')
EMG.visualization(emg_signal,window_size,zooming_from_to)

#After filtration
plt.figure(2)
plt.title('After filtration')
filtered_Data = EMG.filtration(emg_signal,sample_frequency,cutoff_frequency)
EMG.visualization(filtered_Data,window_size,zooming_from_to)

#After RMS
rms_Data=EMG.window_rms(filtered_Data,w_length)
plt.figure(3)
plt.title("After RMS")
EMG.visualization(rms_Data,window_size,zooming_from_to)


#After Envelope
E=EMG.E_try(sample_frequency,rms_Data)
plt.figure(4)
plt.title("Envelope")
EMG.visualization(E,window_size,zooming_from_to)
```
